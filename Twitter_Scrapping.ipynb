```python
#Library to scrap historical tweets with no time period restriction

import GetOldTweets3 as got

import pandas as pd
import numpy as np

#Since people across the globe tweet in different native langaugues, googletrans library can translate all tweets into English

import googletrans
from googletrans import Translator

#Library for language detention
import langdetect
from langdetect import detect


#Extract tweets where the following text is present
search_terms = ['Chrysler']  

tweet_df_all = pd.DataFrame()
for term in search_terms:
    print(term)
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(term).setSince("2020-06-01").setUntil("2020-06-03")

                                           #.setNear("Germany")\
                                           #.setWithin("310mi")\
                                           #.setMaxTweets(1000)
                
    tweet = got.manager.TweetManager.getTweets(tweetCriteria)
    tweet_list = [[tweet[a].id,
                 
               tweet[a].author_id,

               tweet[a].text,
 
               tweet[a].retweets,

               tweet[a].permalink,

               tweet[a].date,

               tweet[a].formatted_date,

               tweet[a].favorites,
 
               tweet[a].mentions,
 
               tweet[a].hashtags,
  
               tweet[a].geo,
 
               tweet[a].urls
  
               ]for a in range(0, len(tweet))]
    
    tweet_df = pd.DataFrame(tweet_list)
    tweet_df['search_term'] = term
    tweet_df_all = tweet_df_all.append(tweet_df)

#If script fails due to too many requests limitation, limit the time period in single query and add multiple search queries and append the data in final dataframe
tweet_df_all.columns = ['id','author_id','text','retweets','permalink','date','formatted_date','favorites','mentions','hashtags','geo','urls','search_term']
tweet_df_all['text'].replace('', np.nan, inplace=True)
tweet_df_all.dropna(subset=['text'], inplace=True)
tweet_df_all.to_csv('Twitter_scrapping_raw_data_Chrysler.csv', index=False)

#To see number of tweets extracted for each search term
tweet_df_all.groupby('search_term')['id'].count()

tweet_df_all['Language_code']=""
count=0 #to bypass records that throw exception
for i in range(0,len(tweet_df_all)):
    try:
        tweet_df_all.iloc[i,13]=detect(tweet_df_all.iloc[i,2])
    except:
        count=count+1

tweet_df_all['Language_code'].replace('', np.nan, inplace=True)
tweet_df_all.dropna(subset=['Language_code'], inplace=True)
tweet_df_all_english= tweet_df_all[tweet_df_all['Language_code']=='en']

#Storing all valid tweets in .csv file
tweet_df_all_english.to_csv('Twitter_scrapping_output_Chrysler.csv', index=False)
```
